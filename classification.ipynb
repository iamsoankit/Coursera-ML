```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#make circular training data
N = 10000
r = 2
x = 2 * r * np.random.rand(1,N) - r
y = 2 * r * np.random.rand(1,N) - r
colors = ((x**2 + y**2)**.5 < 1) + 2
```

```python
plt.scatter(x,y,c=colors,cmap="Blues")
plt.show()
```
